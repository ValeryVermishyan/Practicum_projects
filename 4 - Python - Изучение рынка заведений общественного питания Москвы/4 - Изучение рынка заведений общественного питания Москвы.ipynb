# Описание данных

Файл `/datasets/rest_info.csv` содержит информацию о заведениях общественного питания:
- `name` — название заведения;
- `address` — адрес заведения;
- `district` — административный район, в котором находится заведение, например Центральный административный округ;
- `category` — категория заведения, например «кафе», «пиццерия» или «кофейня»;
- `hours` — информация о днях и часах работы;
- `rating` — рейтинг заведения по оценкам пользователей в Яндекс Картах (высшая оценка — `5.0`);
- `chain` — число, выраженное `0` или `1`, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки):
   - `0` — заведение не является сетевым;
   - `1` — заведение является сетевым.
- `seats` — количество посадочных мест.

Файл `/datasets/rest_price.csv` содержит информацию о среднем чеке в заведениях общественного питания:
- `price` — категория цен в заведении, например «средние», «ниже среднего», «выше среднего» и так далее;
- `avg_bill` — строка, которая хранит среднюю стоимость заказа в виде диапазона, например:
   - «Средний счёт: 1000–1500 ₽»;
   - «Цена чашки капучино: 130–220 ₽»;
   - «Цена бокала пива: 400–600 ₽».
и так далее;
- `middle_avg_bill` — число с оценкой среднего чека, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Средний счёт»:
   - Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
   - Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
   - Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт.
- `middle_coffee_cup` — число с оценкой одной чашки капучино, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Цена одной чашки капучино»:
   - Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
   - Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
   - Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», то в столбец ничего не войдёт.

# 3. Исследовательский анализ данных

Изучим, какие категории заведений представлены в данных. 
Перед началом анализа объявим функцию для создания столбчатых диаграмм

```python
def bar_chart(df, category, figsize_1, figsize_2, x_fontsize, height):

    """
    df - датафрейм, который будет сгруппирован
    category - столбец, по которому произойдет группировка
    figsize_1 - ширина диаграммы
    figsize_2 - высота диаграммы
    x_fontsize - размер шрифта обозначений на оси x
    height - высота долей над соответствующим столбцом
    """
    
    # Группируем датафрейм по категориям
    df_category = df[category].value_counts()
    
    # Создаём контейнер графика matplotlib и задаём его размер. Также задаем шрифт оси X
    plt.figure(figsize=(figsize_1, figsize_2))
    plt.xticks(fontsize=x_fontsize)
    
    # Строим столбчатую диаграмму
    sns.barplot(x=df_category.index, y=df_category.values)

    # Добавляем сетку графика
    plt.grid(axis='y')
    
    # Добавляем на график доли
    total = df_category.sum()
    for i, count in enumerate(df_category.values):
        plt.text(i, count + height, f'{round(count / total*100),2}%', ha='center')
```

```python
# Строим столбчатую диаграмму заведений по категориям
bar_chart(df_cleaned_2, 'category', figsize_1=9, figsize_2=6, x_fontsize=7, height=35)

# Настраиваем оформление графика
plt.title('Кол-во объектов общественного питания в каждой категории')
plt.xlabel('Категория')

# Показываем график
plt.show()
```

- **Кафе** и **рестораны** составляют больше половины всех заведений (28% и 24%)
- Наименьшую долю составляю **столовые** и **булочные** (4% и 3%). Предполагаю, что это связано с падением популярности таких заведений: современный среднестатистический житель такого мегаполиса как Москва не ассоциирует себя с такими заведениями и предпочтет что-то более современное.

  Однако это не значит, что нет смысла открывать такое заведение. Возможно эта ниша "занята несовременными управленцами" и если адаптировать формат под потребителя, то человек может получить совершенно новый интересный для себя опыт и заведение станет успешным. В ходе исследования был изучен рынок общественного питания Москвы и проанализированы данные о 8400 различных заведениях.

**По результатам исследования можно сделать следующие выводы:**

- Наиболее привлекательными **округами** для открытия заведения являются **ЦАО и ЗАО** с самым высоким средним чеком **1142 и 1039**. Однако важно учитывать, что в **ЦАО** очень высокая конкуренция: почти треть всех заведений находится там.
- Наиболее популярные заведения: **кафе и рестораны**. Однако стоит присмотреться к наименее популярным **столовым и булочным**: можно попробовать переосмыслить формат под современные реалии и стать новатором на рынке
- Также важно, что среди категорий сетевыми чаще всего являются **булочная, кофейня, пиццерия**, поэтому возможно будет сложно бороться с конкуренцией в этих категориях
- Чем выше рейтинг заведения, тем выше ценовая категория

**Рекомендации:**

- В качестве округа для открытия заведения рекомендую присмотреться к ЗАО, так как там высокий средний чек и невысокая конкуренция.
- В качестве категории заведения рекомендую рассмотреть **кафе или ресторан**. Либо можно рискнуть и попробовать открыть **столовую или булочную** в новом формате
- Открывая ресторан, следует ориентироваться на большое кол-во посадочных мест - примерно 100
- Следует уделять внимания качеству услуг и сервису, так как высокое качество дает возможность выставлять более высокие цены
